## Package Installation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Loading first dataset

```python
phu_dataset = pd.read_csv('https://data.ontario.ca/dataset/5472ffc1-88e2-48ca-bc9f-4aa249c1298d/resource/36048cc1-3c47-48ff-a49f-8c7840e32cc2/download/ongoing_outbreaks_phu.csv')
phu_dataset
```

## Loading second dataset

```python
ongoing_outbreaks = pd.read_csv('https://data.ontario.ca/dataset/5472ffc1-88e2-48ca-bc9f-4aa249c1298d/resource/66d15cce-bfee-4f91-9e6e-0ea79ec52b3d/download/ongoing_outbreaks.csv')
ongoing_outbreaks
```

## Obtain how many outbreaks are in each different PHU per day in the different Public Health Units

```python
phu_dataset_by_date_phu_name = phu_dataset.pivot_table(index='date', columns='phu_name', values='number_ongoing_outbreaks', aggfunc='sum')
phu_dataset_by_date_phu_name
```


## Plot the different PHUs vs the number of ongoing outbreaks 

```python
phu_outbreaks = phu_dataset[['phu_name','number_ongoing_outbreaks']]
phu_pivot = phu_outbreaks.pivot_table(index='phu_name', aggfunc='sum').sort_values(by='number_ongoing_outbreaks', ascending=True)

print('Maximum value of outbreaks:',phu_pivot.values.max())
print('Minimum value of outbreaks:',phu_pivot.values.min())
phu_pivot.plot(kind='bar')
plt.xlabel('PHU Name', fontsize=15)
plt.ylabel('Total Amount of ongoing outbreaks', fontsize=15)
plt.title('Public Health Unit vs Ongoing Outbreaks', fontsize=18)
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

## Per date get the number of ongoing outbreaks by the 7 different groups

```python
outgoing_outbreaks_by_date_group = ongoing_outbreaks.pivot_table(index='date', columns='outbreak_group', values='number_ongoing_outbreaks', aggfunc='sum')
outgoing_outbreaks_by_date_group
```

## Obtain the number of ongoing outbreaks depending on its different subgroup within each different group

```python
ongoing_outbreaks_subgroup = ongoing_outbreaks[['outbreak_group','outbreak_subgroup', 'number_ongoing_outbreaks']]
ongoing_outbreaks_pivot = ongoing_outbreaks_subgroup.pivot_table(index=['outbreak_group','outbreak_subgroup'], aggfunc='sum')
ongoing_outbreaks_pivot
```

## Plot of the total of ongoing outbreaks in each group 

```python
ongoing_outbreaks_group = ongoing_outbreaks[['outbreak_group', 'number_ongoing_outbreaks']]
ongoing_outbreaks_group_pivot = ongoing_outbreaks.pivot_table(index=['outbreak_group'], aggfunc='sum').sort_values(by='number_ongoing_outbreaks')

ongoing_outbreaks_group_pivot.plot(kind='barh')
plt.xlabel('Total Amount of ongoing outbreaks', fontsize=15)
plt.ylabel('Outbreak Group', fontsize=15)
plt.title('Outbreak Group vs Ongoing Outbreaks', fontsize=20)
plt.grid(True)
plt.show()
```

## For each subgroup its respective ongoing outbreaks

```python
ongoing_outbreaks_subgroup_pivot = ongoing_outbreaks_subgroup.pivot_table(index='outbreak_subgroup', aggfunc='sum').sort_values(by='number_ongoing_outbreaks', ascending=False)

ongoing_outbreaks_subgroup = ongoing_outbreaks_subgroup.sort_values(by='number_ongoing_outbreaks', ascending=False)
ongoing_outbreaks_subgroup_pivot.plot(kind='bar')
plt.xlabel('Outbreak Subgroup', fontsize=15)
plt.ylabel('Total Amount of ongoing outbreaks', fontsize=12)
plt.title('Outbreak Subgroup vs Ongoing Outbreaks', fontsize=16)
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```

## Obtain the outbreaks in Toronto in three different years

```python
phu_Toronto_2020 = phu_dataset_by_date_phu_name['TORONTO']['2020-11-01':'2020-12-31'].values
phu_Toronto_2021 = phu_dataset_by_date_phu_name['TORONTO']['2021-01-01':'2021-12-31'].values
phu_Toronto_2022 = phu_dataset_by_date_phu_name['TORONTO']['2022-01-01':'2022-12-31'].values

print('The maximum value of outbreaks in the year 2020 was:', phu_dataset_by_date_phu_name['TORONTO']['2020-11-01':'2020-12-31'].values.max())
print('The maximum value of outbreaks in the year 2021 was:', phu_dataset_by_date_phu_name['TORONTO']['2021-01-01':'2021-12-31'].values.max())
print('The maximum value of outbreaks in the year 2022 was:', phu_dataset_by_date_phu_name['TORONTO']['2022-01-01':'2022-12-31'].values.max())

plt.plot(phu_Toronto_2020,label='Outbreaks in Toronto 2020')
plt.plot(phu_Toronto_2021, label='Outbreaks in Toronto 2021')
plt.plot(phu_Toronto_2022, label='Outbreaks in Toronto 2022')
plt.xlabel('Days', fontsize=15)
plt.ylabel('Number of Outbreaks', fontsize=15)
plt.title('Days vs Number of Outbreaks in Toronto', fontsize=18)
plt.grid(True)
plt.legend()
plt.show()
```
